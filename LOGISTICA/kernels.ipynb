```python
import seaborn as sns

import matplotlib.pyplot as plt

plt.figure(figsize=(15, 15), dpi=80)
sns.heatmap(df.corr(), annot=True)
```

```python
df.columns
```

```python
from sklearn.linear_model import LogisticRegression
Features=[
'RevolvingUtilizationOfUnsecuredLines', 'Age',
       'NumberOfTime30-59DaysPastDueNotWorse', 'DebtRatio', 'MonthlyIncome',
       'NumberOfOpenCreditLinesAndLoans', 'NumberOfTimes90DaysLate',
       'NumberRealEstateLoansOrLines', 'NumberOfTime60-89DaysPastDueNotWorse',
       'NumberOfDependents'
       ]
X = df[Features]
y = df["SeriousDlqin2yrs"]

X.head()
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=30)

```

```python
y_train.shape
```

```python



from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV




knn = KNeighborsClassifier(n_neighbors=5)
parameters = {
    'kernel': ['linear', 'rbf', 'sigmoid', 'poly'],
    'C': [0.001, 0.1, 0.5, 1, 5, 10, 100],
    'degree': [1,2,3,4,5,6,7],
    'gamma': ['scale', 'auto']


    
}

['algorithm', 'leaf_size', 'metric', 'metric_params',  'n_neighbors', 'p', 'weights']


clf = GridSearchCV(estimator = knn,
                  param_grid = parameters,
                  n_jobs = -1,
                  cv = 10,
                  scoring="roc_auc_ovo")
clf.fit(X_train, y_train)

y_proba = knn.predict_proba(X_test)[:, 1]
y_pred = (y_proba > 0.5).astype(int)

print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
acierto = accuracy_score(y_test, y_pred)

error = 1 - acierto
print("Acierto:", round(acierto*100, 2), "%")
print("Error:", round(error*100, 2), "%")
from sklearn.metrics import confusion_matrix



sns.heatmap(confusion_matrix(y_test, y_pred, normalize='true'), annot=True, 
            fmt='.2%', cmap='Blues');
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1:", f1_score(y_test, y_pred))
print("AUC-ROC:", roc_auc_score(y_test, y_pred))
print("score train:", knn.score(X_train, y_train))
print("score test:", knn.score(X_test, y_test))
```

```python
from sklearn.model_selection import cross_validate
cross_validate(knn, X_train, y_train, cv=5)["test_score"].mean()
```
