```python
import seaborn as sns

import matplotlib.pyplot as plt

plt.figure(figsize=(15, 15), dpi=80)
sns.heatmap(df.corr(), annot=True)
```

```python
df.columns
```

```python
from sklearn.linear_model import LogisticRegression
Features=[
       'RevolvingUtilizationOfUnsecuredLines', 
       'NumberOfTime30-59DaysPastDueNotWorse', 
       'NumberOfOpenCreditLinesAndLoans', 'NumberOfTimes90DaysLate',
       'NumberOfTime60-89DaysPastDueNotWorse',
       ]
X = df[Features]
y = df["SeriousDlqin2yrs"]

X.head()
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=30)


```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=10000, class_weight='balanced', )
model.fit(X_train, y_train)

y_proba = model.predict_proba(X_test)[:, 1]
y_pred = (y_proba > 0.4).astype(int)

print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, classification_report
acierto = accuracy_score(y_test, y_pred)

error = 1 - acierto
print("Acierto:", round(acierto*100, 2), "%")
print("Error:", round(error*100, 2), "%")
from sklearn.metrics import confusion_matrix



sns.heatmap(confusion_matrix(y_test, y_pred, normalize='true'), annot=True, 
            fmt='.2%', cmap='Blues');
print(classification_report(y, model.predict(X), zero_division=0))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1:", f1_score(y_test, y_pred))
print("AUC-ROC:", roc_auc_score(y_test, y_pred))
print("score train:", model.score(X_train, y_train))
print("score test:", model.score(X_test, y_test))
```

```python
from sklearn import metrics

y_pred_proba = model.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
from sklearn.model_selection import cross_validate
cross_validate(model, X_train, y_train, cv=5)["test_score"].mean()
```

```python
from sklearn.model_selection import cross_validate
cross_validate(model, X_train, y_train, cv=5)["test_score"].mean()
from sklearn.calibration import calibration_curve



fraccion_positivos, media_prob_predicha = calibration_curve(y_test, y_proba, n_bins=20)
import matplotlib.pyplot as plt
fig, axs = plt.subplots(nrows=2, ncols=1,figsize=(6, 2*3.84))

axs[0].plot(media_prob_predicha, fraccion_positivos, "s-", label="logistica")
axs[0].plot([0, 1], [0, 1], "k:", label="Calibraci√≥n perfecta")
axs[0].set_ylabel("Proporci√≥n de clasificaci√≥n correcta")
axs[0].set_xlabel("Probabilidad media estimada por el modelo (predict_proba)")
axs[0].set_title('Curva de calibrado (reliability curve)')
axs[0].legend()

axs[1].hist(y_proba, range=(0, 1), bins=10, density=True, lw=2, alpha = 0.3)
axs[1].set_xlabel("Probabilidad estimada por el modelo (predict_proba)")
axs[1].set_ylabel("Count")
axs[1].set_title('Distribuci√≥n de las probabilidades predichas por el modelo')

plt.tight_layout()
plt.show();
```
