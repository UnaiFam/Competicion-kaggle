```python
import seaborn as sns

import matplotlib.pyplot as plt

plt.figure(figsize=(15, 15), dpi=80)
sns.heatmap(df.corr(), annot=True)
```

```python
from sklearn.linear_model import LogisticRegression
Features=[
'RevolvingUtilizationOfUnsecuredLines', 'Age',
       'NumberOfTime30-59DaysPastDueNotWorse', 'DebtRatio', 'MonthlyIncome',
       'NumberOfOpenCreditLinesAndLoans', 'NumberOfTimes90DaysLate',
       'NumberRealEstateLoansOrLines', 'NumberOfTime60-89DaysPastDueNotWorse',
       'NumberOfDependents']
X = df[Features]
y = df["SeriousDlqin2yrs"]

X.head()
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=42)


```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate
tree_clf = DecisionTreeClassifier(random_state=50,  max_depth= 39, criterion="entropy", min_samples_leaf=6)
tree_clf.fit(X_train, y_train)

tree_clf.score(X_train,y_train)
print("profundidad", tree_clf.get_depth())
print("score train", (tree_clf.score(X_train,y_train)))
print("cross_val", (cross_validate(tree_clf, X_train, y_train, cv=5)["test_score"].mean()))
print("score test", (tree_clf.score(X_test,y_test)))
from sklearn.metrics import  mean_squared_error
import numpy as np
y_pred=(tree_clf.predict_proba(X_test)[:,1] > 0.34).astype(int)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("rmse", rmse)
```

```python
from sklearn.metrics import accuracy_score
acierto = accuracy_score(y_test, y_pred)

error = 1 - acierto
print("Acierto:", round(acierto*100, 2), "%")
print("Error:", round(error*100, 2), "%")
from sklearn.metrics import confusion_matrix

sns.heatmap(confusion_matrix(y_test, y_pred, normalize='true'), annot=True, 
            fmt='.2%', cmap='Blues');
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1:", f1_score(y_test, y_pred))
print("AUC-ROC:", roc_auc_score(y_test, y_pred))
```
