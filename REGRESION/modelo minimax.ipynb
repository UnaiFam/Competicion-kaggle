```python
import pandas as pd
df=pd.read_csv("df_train.csv")
s=df.columns
df
```

```python
import sklearn.preprocessing 
import seaborn as sns
df_scale=sklearn.preprocessing.minmax_scale(df)
df_scale
```

```python
df=pd.DataFrame(df_scale, columns=s)
df
```

```python
Feeatures=['Sys_enc', 'HD',
       'Touch', 'IPS', '4K', 'Screen_Enc',  'SSD',  
        'CPU_enc', 'GPU_enc']
#X=df[Feeatures]
X=df.drop("Price_euros", axis=1)
y=df["Price_euros"]
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 24)
lm=LinearRegression()
lm.fit(X_train, y_train)


prediction_lm=lm.predict(X_test)
sns.scatterplot(x=y_test, y=(prediction_lm))
from sklearn.metrics   import mean_squared_error
from  sklearn.model_selection import cross_validate
import numpy as np
rmsec_lm=np.sqrt(mean_squared_error(pd.DataFrame(y_test),pd.DataFrame(prediction_lm)))



print("R2 train", lm.score(X_train, y_train))
print("R2 test",lm.score(X_test, y_test))
print("cross_val", cross_validate(lm, X_train, y_train, cv=5)["test_score"].mean())
print(rmsec_lm)
```
