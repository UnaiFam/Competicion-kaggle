```python
import re
#hago una funcion para sacar memoria

def parse_memory(mem_str):
    #defino patron 
    #el patron siemte es un numero| con coma| segifo de memoria| esmacio en blanco | str de letars ilimintadas tipo de memoria
    pattern = r'(\d+\.?\d*)\s*(TB|GB)\s*([A-Za-z ]+)'
    # busco ctoso en el str que le doy, para saber las posiciones
    matches = re.findall(pattern, mem_str)
    mem_dict = {}
    # lo guardo todo en dict
    # saco gb, memoria y tipo
    for amount, unit, mem_type in matches:
        #itero en el diccionarion
        amount_gb = float(amount) * (1024 if unit == 'TB' else 1)
        # si hay tb lo paso a gb 
        mem_type = mem_type.strip().replace(' ', '_')  # Para columnas sin espacios
        # Sumar si ya existe ese tipo
        mem_dict[mem_type] = mem_dict.get(mem_type, 0) + amount_gb
    return mem_dict
# itero para que lea cada  fila
parsed = [parse_memory(s) for s in df_train["Memory"]]
print(parsed)
# Crear DataFrame a partir de lista de diccionarios, las keys serán columnas
df_mem = pd.DataFrame(parsed)
df_mem.fillna(0)
#uno los dataframes
df_train=pd.concat([df_train, df_mem.fillna(0)], axis=1)
```
